````python
num_starts_in_each_line = 10
num_lines = 8
print("Print m x n stars")
_ = [print("*"*num_starts_in_each_line) for i in range(num_lines)]
````

````python
print("Descending with tailing space")
ctr = num_starts_in_each_line
for _ in range(num_lines+1):
    ctr -= 1
    print("*"*ctr)
````

````python
print("Ascending with tailing space")
ctr = 0
for _ in range(num_starts_in_each_line+1):
    ctr += 1
    print("*"*ctr)
    
print("Ascending with tailing space")
ctr = 0
for _ in range(num_starts_in_each_line+1):
    ctr += 1
    leading_spaces = " "*(ctr)
    tailing_spaces = " "*(ctr)
    ctr_str = leading_spaces + "*" + tailing_spaces
    print(ctr_str)        
````

````python
print("\nAscending with leading space\n")
max_num_stars_in_a_line = 10
for l_no in range(1, max_num_stars_in_a_line):
    spaces = " "* (max_num_stars_in_a_line-l_no)
    stars = "*"*l_no
    ctr_str = spaces + stars
    print(ctr_str)
    
print("\nAscending with leading space\n")
max_num_stars_in_a_line = 10
for l_no in range(0, max_num_stars_in_a_line):
    leading_spaces = " "* (max_num_stars_in_a_line-l_no)
    tailing_spaces = " "*(l_no)
    ctr_str = leading_spaces + "*" + tailing_spaces + "*"
    print(ctr_str)
    
print("\nAscending with leading space\n")
max_num_stars_in_a_line = 10
for l_no in range(0, max_num_stars_in_a_line):
    leading_spaces = " "* (max_num_stars_in_a_line-l_no)
    tailing_spaces = " "*(l_no)
    ctr_str = leading_spaces + "*"
    print(ctr_str)
    
print("\nAscending with leading space\n")
max_num_stars_in_a_line = 10
for l_no in range(0, max_num_stars_in_a_line):
    if l_no != (max_num_stars_in_a_line-1):
        leading_spaces = " "* (max_num_stars_in_a_line-l_no)
        tailing_spaces = " "*(l_no)
        ctr_str = leading_spaces + "*" + tailing_spaces + "*"
    else:
        ctr_str = " "+" *"*(l_no-4)
    print(ctr_str)
````

````python
max_num_starts_in_a_line = 20

total_num_lines = 2 * max_num_starts_in_a_line - 1

print(f"max_num_starts_in_a_line: {max_num_starts_in_a_line}\ntotal_num_lines: {total_num_lines}")

print("Arrow Head Stars Pointing Left")

ctr = 1
for l_no in range(total_num_lines):
    if l_no < max_num_starts_in_a_line:
        spaces = " "*(max_num_starts_in_a_line-l_no-1)
        stars = "*"*(l_no+1)
    else:
        spaces = " "*(l_no - max_num_starts_in_a_line+1)
        stars = "*"*(max_num_starts_in_a_line-len(spaces))
    ctr_str = spaces + stars        
    print(ctr_str)
````

````python
size_of_squeare = 30

print("Square of Size:",size_of_squeare)

for l_no in range(size_of_squeare):
    if l_no == 0 or l_no == (size_of_squeare-1):
        print(" *"*(size_of_squeare))
    else:
        p_str = "*"+"  "*(size_of_squeare)+"*"
        print(p_str)
````

````python
rect_len, rect_wid = 15, 40

print("Rectangle of shape:",rect_len,"X",rect_wid)

for l_no in range(rect_len):
    if l_no == 0 or l_no == (rect_len-1):
        print(" *"*(rect_wid))
    else:
        p_str = "* "+"  "*(rect_wid-1)+"*"
        print(p_str)
````

````python
rhomb_size = 5
print("Rhombous of size:",rhomb_size)

num_lines_for_rhomb = (rhomb_size * 2) - 1

for l_no in range(num_lines_for_rhomb+2):
    if l_no <= rhomb_size:
        spaces = " "*(rhomb_size-l_no)
        stars = "*"*((l_no*2) + 1)
        p_str = spaces+stars+spaces
    else:
        spaces = " "*(l_no - rhomb_size)
        stars = "*"*(((rhomb_size-len(spaces))*2)+1)
        p_str = spaces+stars+spaces
    print(p_str)
````
